Use CNN to do handwritten digits classification using MNIST dataset.Check how accuracy improves fast with CNN and figure out how CNN can be a better choice for doing image classification compared to ANN.

```python
import tensorflow as tf
from tensorflow.keras import datasets,layers, models
import matplotlib.pyplot as plt
import numpy as np
```

Load the MNIST dataset

```python
mnist = tf.keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

Normalize pixel values to be between 0 and 1

```python
train_images, test_images = train_images / 255.0, test_images / 255.0
```

Build the model

```python
ann = models.Sequential()
```

Flatten layer to convert 28x28 images to a flat vector

```python
ann.add(layers.Flatten(input_shape=(28, 28)))
```

Experiment with different architectures and activation functions

```python
ann.add(layers.Dense(128, activation='relu'))
ann.add(layers.Dense(64, activation='relu'))
ann.add(layers.Dense(10, activation='softmax'))
```

Compile the model

```python
ann.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

Train the ANN Model

```python
ann.fit(train_images, train_labels, epochs=10, validation_data=(test_images, test_labels))
```

Evaluate ANN Model

```python
ann.evaluate(test_images, test_labels)
```

Predict ANN Model

```python
predictions = ann.predict(test_images)
predictions
```

CNN Model

```python
cnn = models.Sequential([
    layers.Conv2D(filters=28, kernel_size=(3,3),activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(filters=64, kernel_size=(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
cnn.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
cnn.fit(train_images, train_labels, epochs=10, validation_data=(test_images, test_labels))
```

Evaluate CNN Model

```python
cnn.evaluate(test_images, test_labels)
```

Prediction CNN Model

```python
predictions = cnn.predict(test_images)
predictions
```

```python
from sklearn.metrics import confusion_matrix , classification_report
import numpy as np
```

ANN Classification Report

```python
y_pred = ann.predict(test_images)
y_pred_classes = [np.argmax(element) for element in y_pred]
```

```python
print("Classification Report :\n" , classification_report(test_labels,y_pred_classes))
```

CNN Classification Report

```python
y_pred = cnn.predict(test_images)
y_pred_classes = [np.argmax(element) for element in y_pred]
```

```python
print("Classification Report :\n" , classification_report(test_labels,y_pred_classes))
```
